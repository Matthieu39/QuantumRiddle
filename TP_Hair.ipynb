```python
# We have 4 people, positioned in a line, whose hair has been dyed. Two of them have now blue hair, the other two orange hair, 
# but none of them know their own color. We must find a way for them to determine it with precision without moving.
# We're gonna use the fact that the number of each color are even, and the observations of each of them.
```

```python
# Imports
```

```python
import numpy as np
from qiskit import *
from qiskit.visualization import *
```

```python
# Resolution
```

```python
circuit = QuantumCircuit(8)
circuit.draw('mpl')
```

```python
circuit.h(0)
circuit.h(1)
circuit.h(2)
circuit.h(3)
circuit.barrier(0,1,2,3,4,5,6,7)
circuit.draw('mpl')
```

```python
# The person in last checks if the number of blue hair in the line is odd or even and can determine the color of their own
```

```python
circuit.cx(1, 4)
circuit.cx(2, 4)
circuit.cx(3, 4)
circuit.draw('mpl')
```

```python
# Everyone is informed of the answer
```

```python
circuit.barrier(0, 1, 2, 3, 4, 5, 6, 7)
circuit.cx(4, 5)
circuit.cx(4, 6)
circuit.cx(4, 7)
circuit.draw('mpl')
```

```python
# The second person checks if the hair of the last two are odd or even and can determine the color of their own
```

```python
circuit.barrier(0, 1, 2, 3, 4, 5, 6, 7)
circuit.cx(2, 5)
circuit.cx(3, 5)
circuit.draw('mpl')
```

```python
# Everyone is informed of the answer
```

```python
circuit.barrier(0, 1, 2, 3, 4, 5, 6, 7)
circuit.cx(5, 6)
circuit.cx(5, 7)
circuit.draw("mpl")
```

```python
# The third person checks the hair of the last one and can determine the color of their own
```

```python
circuit.barrier(0, 1, 2, 3, 4, 5, 6, 7)
circuit.cx(3, 6)
circuit.draw("mpl")
```

```python
# Everyone is informed of the answer
```

```python
circuit.barrier(0, 1, 2, 3, 4, 5, 6, 7)
circuit.cx(6, 7)
circuit.draw("mpl")
```

```python
circuit.measure_all()
circuit.draw("mpl")
```
