```python
# We're facing two doors, each protected by a guard. There's a treasure behind one of them. 
# We know one of the two guards is a liar and whe should determine how to ask them which door is the one we want
```

```python
# Imports
```

```python
from qiskit import *
from qiskit import Aer
from qiskit.visualization import plot_state_city
```

```python
# Resolution
# We create 3 qubits : 2 Guards / 1 liar
```

```python
circuit = QuantumCircuit(3)
circuit.draw('mpl')
```

```python
circuit.h(0)
circuit.cx(0,1)
circuit.h(2)
circuit.draw('mpl')
```

```python
circuit.cx(2,1)
circuit.x(2)
circuit.cx(2,0)
circuit.x(2)
circuit.draw('mpl')
```

```python
circuit.swap(0, 1)
circuit.x(0)
circuit.x(1)
circuit.draw('mpl')
```

```python
circuit.cx(2,1)
circuit.x(2)
circuit.cx(2,0)
circuit.x(2)
circuit.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator')
job = backend.run(circuit)
res = job.result()
output = res.get_statevector(circuit, decimals=3)
plot_state_city(output)
```
